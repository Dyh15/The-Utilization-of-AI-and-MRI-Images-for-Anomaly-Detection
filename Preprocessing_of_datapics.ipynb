```python
from google.colab import files
from zipfile import ZipFile
import cv2
import os
import numpy as np
from PIL import Image
import shutil
```

```python
file1='data_0.zip'
file2='data_1.zip'
with ZipFile(file1,'r') as zip_ref:
  zip_ref.extractall('data_0')
with ZipFile(file2,'r') as zip_ref:
  zip_ref.extractall('data_1')
print('done')
```

```python
path_data_0= '/content/data_0/data_0'
new_path_0= '/content/preprocessed_data_0'
path_data_1= '/content/data_1/data_1'
new_path_1= '/content/preprocessed_data_1'
```

```python
image_files_0= os.listdir(path_data_0)
image_files_1= os.listdir(path_data_1)
size=(350, 350)
```

```python
i=0
for filename in image_files_0:
    input_path = os.path.join(path_data_0, filename)
    image = cv2.imread((input_path))
    try:
      resized = cv2.resize(image, size)
    except:
      break
    filtered1 = cv2.medianBlur(resized, 3)
    mask = cv2.subtract(resized, filtered1)
    filtered2 = cv2.add(resized, mask)
    output_path = os.path.join(new_path_0, filename)
    cv2.imwrite(output_path, filtered2)
    i=i+1
i
```

```python
i=0
for filename in image_files_1:
    input_path = os.path.join(path_data_1, filename)
    image = cv2.imread((input_path))
    try:
      resized = cv2.resize(image, size)
    except:
      break
    filtered1 = cv2.medianBlur(resized, 3)
    mask = cv2.subtract(resized, filtered1)
    filtered2 = cv2.add(resized, mask)
    output_path = os.path.join(new_path_1, filename)
    cv2.imwrite(output_path, filtered2)
    i=i+1
i
```

```python
shutil.make_archive('preprocessed_data_0', 'zip', '/content/preprocessed_data_0')
files.download('preprocessed_data_0.zip')
```
